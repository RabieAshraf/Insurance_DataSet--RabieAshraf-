```python
male_no = df['sex'].value_counts()['male']
male_no
```

```python
female_no = df['sex'].value_counts()['female']
female_no
```

```python
plt.pie([male_no,female_no], labels=['male','female'], autopct='%1.1f%%') 
```

```python
df_male = df.loc[df['sex']=='male']
df_male
```

```python
df_female = df.loc[df['sex'] == 'female']
df_female
```

```python
smoke_men = df_male['smoker'].value_counts()['yes']
smoke_men 
```

```python
smoke_fem = df_female['smoker'].value_counts()['yes']
smoke_fem
```

```python
plt.pie([smoke_men,smoke_fem], labels=['men','women'], autopct='%1.1f%%') #Men smoke more than women
```

```python
#compare smokers and bmi
sns.barplot(data=df, x='smoker', y='bmi') #Same value
```

```python
#We can conclude from these data the following
#Men had more charges than women -> may aim for more men to get more charges
#Smokers had very high charges
#The southeast region had most charges
#Children didn't play an important role in charges
#Bmi didn't have a high effect in charges
```
